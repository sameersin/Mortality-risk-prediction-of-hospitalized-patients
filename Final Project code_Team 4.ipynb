```python
def correlation(dataset, threshold):
    correlated_pairs = []  # List to hold pairs of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:  # Check if correlation exceeds threshold
                colname1 = corr_matrix.columns[j]  # Name of the first column in the pair
                colname2 = corr_matrix.columns[i]  # Name of the second column in the pair
                correlated_pairs.append((colname1, colname2))  # Append the tuple to the list
    return correlated_pairs

# Usage example with a DataFrame 'data' and a threshold for strong correlation
strong_correlations = correlation(data, 0.7)
for i in strong_correlations:
    print(i)
```
